```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.morphology import disk
from skimage.filters import rank


image_path = '/content/drive/MyDrive/OIP (3).jpg'
image = ski.io.imread(image_path, as_gray=True)
footprint = disk(20)

percentile_result = rank.mean_percentile(image, footprint=footprint, p0=0.1, p1=0.9)
bilateral_result = rank.mean_bilateral(image, footprint=footprint, s0=500, s1=500)
normal_result = rank.mean(image, footprint=footprint)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), sharex=True, sharey=True)
ax = axes.ravel()

titles = ['Original', 'Percentile mean', 'Bilateral mean', 'Local mean']
imgs = [image, percentile_result, bilateral_result, normal_result]
for n in range(0, len(imgs)):
    ax[n].imshow(imgs[n], cmap=plt.cm.gray)
    ax[n].set_title(titles[n])
    ax[n].axis('off')

plt.tight_layout()
plt.show()
```

```python
from skimage.filters import gaussian
image_path = '/content/drive/MyDrive/OIP (3).jpg'
image = ski.io.imread(image_path, as_gray=True)

sigma = 2
gaussian_result = gaussian(image, sigma=sigma)

# Display the original and filtered images
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), sharex=True, sharey=True)
ax = axes.ravel()

titles = ['Original', 'Gaussian Filtered']
imgs = [image, gaussian_result]

for n in range(len(imgs)):
    ax[n].imshow(imgs[n], cmap=plt.cm.gray)
    ax[n].set_title(titles[n])
    ax[n].axis('off')

plt.tight_layout()
plt.show()
```

```python
from skimage.filters.rank import median
image_path = '/content/drive/MyDrive/OIP (3).jpg'
image = ski.io.imread(image_path, as_gray=True)

footprint = disk(5)
median_result = median(image, footprint=footprint)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), sharex=True, sharey=True)
ax = axes.ravel()

titles = ['Original', 'Median Filtered']
imgs = [image, median_result]

for n in range(len(imgs)):
    ax[n].imshow(imgs[n], cmap=plt.cm.gray)
    ax[n].set_title(titles[n])
    ax[n].axis('off')

plt.tight_layout()
plt.show()
```
